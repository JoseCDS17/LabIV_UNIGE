```python
#import necessary modules
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['font.family'] = 'CM Roman'
matplotlib.rcParams['font.size'] = 12
```

```python
with open("./PMT_data.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

min = np.array([float(line.split(',')[0]) for line in lines])
integral = np.array([float(line.split(',')[1]) for line in lines])
print(f'Number of waveforms: {len(min)}')
```

```python
discrimination_value = -0.1
index_dis = np.where(min < discrimination_value)[0]

#efficiency
eff = len(index_dis)/len(min)
print(f'Efficiency: {eff*100:0.2f}%')

#discrimination
min_dis = min[index_dis]
int_dis = integral[index_dis]
```

```python
bins = 50
#Figure
fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(14)

ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

ax1.hist(min_dis,bins = bins, color='b')
ax1.set_xlabel('Voltage amplitude (V)')
ax1.set_ylabel('Counts')
ax1.set_title('Amplitude histogram')

ax2.hist(int_dis,bins = bins, color='r')
ax2.set_xlabel('Waveform integral')
ax2.set_ylabel('Counts')
ax2.set_title('Charge histogram')

fig.savefig('../Images/PMT_AmplitudeCharge.pdf')
```
