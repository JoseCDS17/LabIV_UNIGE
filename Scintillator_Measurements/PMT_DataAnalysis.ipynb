```python
#import necessary modules
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['font.family'] = 'CM Roman'
matplotlib.rcParams['font.size'] = 12
```

```python
with open("./txt_data/SiPMT_data_scMirror.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

max = np.array([float(line.split(',')[0]) for line in lines])
integral = np.array([float(line.split(',')[1]) for line in lines])
print(f'Number of waveforms: {len(max)}')
```

```python
bins = 50
#Figure
fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(14)

ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

ax1.hist(max,bins = bins)
ax1.set_xlabel('Voltage amplitude (V)')
ax1.set_ylabel('Counts')
ax1.set_title('Amplitude histogram')

ax2.hist(integral[integral > 0.2e-9],bins = bins)
ax2.set_xlabel('Waveform integral')
ax2.set_ylabel('Counts')
ax2.set_title('Charge histogram')
```
